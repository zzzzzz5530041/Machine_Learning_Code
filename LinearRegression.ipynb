```python
import numpy as np
import random
import numpy as np
import matplotlib
from matplotlib import pyplot
from matplotlib import pyplot as plt


features = np.array([1,2,3,5,6,7])
labels = np.array([155, 197, 244, 356, 407, 448])



def linear_regression(learning_rate=0.01,epochs=3000):
    slop = random.random()
    bias = random.random()
    errors = []
    for epoch in range(epochs):
        predictions = features[0]*slop+bias
        errors.append(rmse(labels, predictions))
        i  = random.randint(0,len(features)-1)
        draw_line(slop, bias, starting=0, ending=8)
        num_room=features[i]
        price = labels[i]
        slop,bias=squre_trick(slop,bias,learning_rate,price,num_room)
    draw_line(slop, bias,color='red', starting=0, ending=8)
    plot_points(features, labels)
    plt.show()
    plt.scatter(range(len(errors)), errors)
    plt.show()
    return slop,bias


'''
    caculate slop and bias by squre trick
'''
def squre_trick(slop,bias,learning_rate,price,num_room):
    predicted_price = slop*num_room+bias
    slop += learning_rate * num_room*(price - predicted_price) ## m'=m+eta*x*(p-p^)
    bias += learning_rate*(price - predicted_price) ## b'=b+eta*(p-p^)
    return slop,bias

  

def predict_price():
    slop,bias = linear_regression()
    num_rooms = 4
    predict_price = slop*4+bias
    return predict_price
predict_price()



### Error Function
# Root Mean Squared Error
# 均方根误差，RMSE）是衡量模型预测值与实际观测值之间差异的一种常用方法。
#它表示观测值与估计值之间偏差的平方的均值的平方根。RMSE 越小，表示模型的预测精度越高。
# 公式 rmse = 根号下(y-y^)平方的加和/N
# 这个公式计算了 预测值与实际值之间差的平方的平均值，然后取平方根，得到的结果就是均方根误差。
def rmse(labels, predictions):
    n = len(labels) ## number
    differences = np.subtract(labels, predictions) ##极差  真实值和预计值的差
    return np.sqrt(1.0/n * (np.dot(differences, differences)))
    
#np.dot(differences, differences)：计算differences数组的点积，即先对differences数组中的每个元素进行平方，然后将所有平方值相加，得到误差平方和。
# np.sqrt 开根号







```
